```python
# how many listingsare there per state?
count_listing = census_data['STATE'].value_counts().sort_values(ascending = False)
count_listing
```

```python
import matplotlib.pyplot as plt

count_listing = census_data['STATE'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(10,8)) # set figure size
plt.barh(count_listing.index, count_listing.values)

# add data labels
for index, value in enumerate(count_listing.values):
    plt.text(value, index, str(value))

plt.xlabel('Count')
plt.ylabel('State')
plt.title('State-wise Listing Count')

plt.show()
```

```python
# SUMMARY STATISTICS PER STATE AND COUNTRY
census_data.groupby('STATE')['COUNTY'].describe().astype('int')
```

```python
import matplotlib.pyplot as plt

census_data.groupby('STATE')['COUNTY'].describe().astype('int').plot(kind='box', figsize=(8,4))

# add data labels
for i, val in enumerate(census_data.groupby('STATE')['COUNTY'].describe().astype('int').values):
    plt.text(1, i+1, str(val[1]), color='red', fontweight='bold')

plt.ylabel('County')
plt.title('Summary Statistics per State and County')

plt.show()
```

```python
# which City has the most county in it
city_count = census_data.groupby('COUNTY')['REGION'].describe()
city_count
```

```python
import matplotlib.pyplot as plt

# Group the data by city and count the number of counties in each city
city_count = census_data.groupby('COUNTY')['REGION'].describe()

# Sort the cities based on the count of counties in descending order
sorted_city_count = city_count.sort_values(by='count', ascending=False)

# Extract the city names and county counts from the sorted data
cities = sorted_city_count.index
county_counts = sorted_city_count['count'].values

# Create a column chart using matplotlib
plt.figure(figsize=(8,4)) # Adjust the size of the chart as desired
bars = plt.bar(cities, county_counts)

plt.xlabel('City')
plt.ylabel('Number of Counties')
plt.title('Number of Counties in Each City')
plt.xticks(rotation=90)
plt.show()
```

# Observation.
- less than 200 cties have high number of counties
- about half the cities have minimal number of counties

```python
# what region has the most division in it? top 5
div_5reg = census_data.groupby('REGION')['DIVISION'].sum()[:5]
div_5reg
```

```python
import matplotlib.pyplot as plt

# Group the data by division and sum the regions in each division
div_5reg = census_data.groupby('DIVISION')['REGION'].sum()[:5]

# Sort the divisions based on the sum of regions in descending order
sorted_div_5reg = div_5reg.sort_values(ascending=False)

# Extract the division names and region sums from the sorted data
divisions = sorted_div_5reg.index
region_sums = sorted_div_5reg.values

# Create a vertical bar chart using matplotlib
plt.bar(divisions, region_sums)
plt.xlabel('Division')
plt.ylabel('Number of Regions')
plt.title('Top 5 Divisions with the Most Regions')
plt.show()
```
